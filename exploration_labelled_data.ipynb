# Preprocessed data exploration

Now that the data has been labelled, we will do some data exploration and see if we can exploit certain patterns underlying the data. 

```python
# Bootstrap
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import sklearn

LABELS = [
    "Black",
    "Brown",
    "Red",
    "Orange",
    "Yellow",
    "Green",
    "Blue",
    "Purple",
    "Gray",
    "Silver",
    "Gold",
    "Noise"
    ]

FILEPATH_LABEL_CSV = "data/training/"
# df = pd.read_csv(FILEPATH_LABEL_CSV + "labels_2023-06-24_cleaned.csv")
df = pd.read_csv(FILEPATH_LABEL_CSV + "labels_2023-07-24.csv")
```

```python
# Group the DataFrame by the column and count the instances
grouped_df = df.groupby('label').size().reset_index(name='count')

print("Label sizes:")
# Print the sum and name for each group
for index, row in grouped_df.iterrows():
    group_sum = row['count']
    group_name = row['label']
    print(f"{group_name}: {group_sum}")
```

```python
# Plot LAB

# Set up figure and axes
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Iterate over unique labels
# labels = ["Red", "Orange", "Yellow", "Green", "Blue", "Black"]
# for label in labels:
for label in df['label'].unique():
    # Filter DataFrame by label
    label_df = df[df['label'] == label]
    
    axs[0].scatter(label_df['a'], label_df['L'], label=label)
    axs[1].scatter(label_df['b'], label_df['L'], label=label)
    axs[2].scatter(label_df['a'], label_df['b'], label=label)

# Set titles and legends
axs[0].set_title('L')
axs[1].set_title('a')
axs[2].set_title('b')
axs[0].legend()
axs[1].legend()
axs[2].legend()
# Set limits
axs[0].set_ylim((0,100))
axs[1].set_ylim((-128,128))
axs[2].set_ylim((-128,128))
# Set axis labels
# Set x-axis and y-axis labels
axs[0].set_xlabel('a')
axs[0].set_ylabel('L')
axs[1].set_xlabel('b')
axs[1].set_ylabel('L')
axs[2].set_xlabel('a')
axs[2].set_ylabel('b')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 3D Visualization
import plotly.graph_objects as go

fig = go.Figure()

label_colors = {
    'Black' : 'black',
    'Brown' : 'brown',
    'Red' : 'red',
    'Orange' : 'orange',
    'Yellow' : 'yellow',
    'Green': 'green',
    'Blue': 'blue',
    'Purple': 'purple',
    'White' : 'white',
    'Grey': 'grey',
    'Gold': 'gold',
    'Silver': 'silver',
    'Noise': 'hotpink'

}

# Add trace for each label
for label, color in label_colors.items():
    label_df = df[df['label'] == label]
    fig.add_trace(go.Scatter3d(
        x=label_df['L'],
        y=label_df['a'],
        z=label_df['b'],
        mode='markers',
        marker=dict(size=5, color=label_colors[label], opacity=0.8),
        name=label
    ))

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='L', yaxis_title='a', zaxis_title='b'))

# Show the plot
fig.show()
```

```python
# Pearson's coefficient
selected_columns = ['L', 'a', 'b', 'B', 'G', 'R', 'H', 'S', 'V', 'Y', 'Cr', 'Cb']
selected_df = df[selected_columns]
correlation_matrix = selected_df.corr(method='pearson', numeric_only=True).abs()  # Calculate the correlation matrix


n = correlation_matrix.shape[0]  # Number of parameters
# correlation_matrix.values[[np.arange(n)] * 2] = np.nan  # Set diagonals to NaN
correlation_matrix.values[np.tril_indices(n)] = np.nan  # Set lower triangular values to NaN
sorted_corr_matrix = correlation_matrix.unstack().sort_values(ascending=False)
print(correlation_matrix)
print(sorted_corr_matrix[:10])
```
